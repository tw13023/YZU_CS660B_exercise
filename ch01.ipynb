```python
#------------------------------------------------------------
# Practice 1-3
# Let x = [1,2,3] , y = x ** 4, z = 2*y, u = mean(z), 
# Try to generate a computational graph about this relation 
# -----------------------------------------------------------
x = torch.tensor([1,2,3], dtype=torch.float32 ,requires_grad=True)
y = x ** 4
z = 2 * y 
u = z.mean()
u.backward()

make_dot(u)
#dot = make_dot(u)
#dot.view()
```

```python
#------------------------------------------------------------
# Practice 1-4
# Write a program to estimate f'(1), 
# when f(x) = g(x) ** 5 and g(x) = x**2 + 1 
# -----------------------------------------------------------

def dfunc(x):
    x = torch.tensor(x, dtype=torch.float32, requires_grad=True)
    g = x ** 2 + 1 
    f = g ** 5 
    f.backward()
    return x.grad 

print(dfunc(1))


```
